```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips
```

# 4.1 Changing Plot Style and Color

### 4.1.1 Changing the figure style
<ul>
    <li><b>sns.set_style('darkgrid')</b></li> It changes background and axes
    <li><b>Default value is white</b></li>
    <li><b>Other Options are white, dark, whitegrid, darkgrid, ticks</b></li>
</ul>

```python
sns.set_style('darkgrid')
sns.catplot(x='day', y='total_bill', data=tips, kind='point')
```

### 4.1.2 Changing the Palette color
<ul>
    <li><b>sns.set_palette('RdBu')</b></li> Changes the color of the main element of the plot
    <li><b>Diverging Palette</b> - are great to visualize a variable whose two ends are opposite and has a neutral midpoint</li>
    <li><b>Sequential Palette</b> - are great to emphasize a variable that is on a continous scale</li>
    <li><b>Custom Palette</b></li>
</ul>

```python
sns.set_palette('Greys')
sns.catplot(x='day', y='total_bill', data = tips, kind='bar')
```

### 4.1.3 Changing the Scale
<ul>
    <li><b>sns.set_context('paper')</b></li> It changes scale of the plot elements and labels
    <li><b>Smallest to Largest values are:</b> - 'paper', 'notebook', 'talk', 'poster'</li>
    <li><b>Default value is 'paper'</b></li>
</ul>

```python
sns.set_context('poster')
sns.catplot(x='day', y='total_bill', data=tips, kind='point', hue='smoker', join=True, capsize=0.1)
```

# 4.2&4.3 Adding Titles and Labels (Part1 & Part2)

Seaborn plot functions creates two types of plot objects: FacetGrid and AxesSubplot

<ul>
    <li><b>AxesSubplot</b> - created using single plot functions like sns.countplot() or sns.scatterplot()</li>
    <li><b>FacetGrid</b> - created using multiple plot functions like sns.relplot() or sns.catplot()</li>
</ul>

```python
# FACETGRID
g = sns.catplot(x='day', y='total_bill', data=tips, kind='point', col='smoker')
type(g)
g.fig.suptitle('Average Bill', y=1.02)
g.set_titles('For smoker = {col_name}')
g.set(xlabel='Days', ylabel='Total Bill')
plt.xticks(rotation=90) # using matplotlib xticks function to rotate x_tick_labels
```

```python
# AXESSUBPLOT
g = sns.boxplot(x='day', y='total_bill', data=tips)
type(g)
g.set_title('Number of eating in Restaurant')
g.set(xlabel='Days', ylabel='Distribution of Total Bill')
plt.xticks(rotation=30)  # using matplotlib xticks function to rotate x_tick_labels
```
